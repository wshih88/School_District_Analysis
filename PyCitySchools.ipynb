```python
import pandas as pd
import os
```

```python
# Files to load
school_data_to_load = 'Resources/schools_complete.csv'
student_data_to_load = os.path.join('Resources', 'students_complete.csv')
```

```python
# Reading school data and store in Pd df
school_data_df = pd.read_csv(school_data_to_load)
school_data_df.head()
```

```python
# Read student data and store in pd df
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
# Determine if there are any missing values in the school data.
school_data_df.isnull()
```

```python
# Determine if there are any missing values in the student data. .isnull = blank, .notnull = not blank
student_data_df.notnull().sum()
```

```python
# Determine data types for student DataFrame
student_data_df.dtypes
```

```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

# Iterate thru the words in the pre_suff list and replace them w/ an empty space, ''.
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")

print(student_data_df.head)
```
